```python
from transformers import BertForQuestionAnswering, BertTokenizer
import torch

# Load pre-trained BERT model and tokenizer
model = BertForQuestionAnswering.from_pretrained('bert-large-uncased-whole-word-masking-finetuned-squad')
tokenizer = BertTokenizer.from_pretrained('bert-large-uncased-whole-word-masking-finetuned-squad')

# Define context (this could be a conversation or passage)
context = """
In the year 2024, natural language processing systems have advanced significantly, allowing machines to not only understand human language but also generate coherent and contextually appropriate responses.
With the rise of transformer-based models like BERT, tasks like sentiment analysis, machine translation, and question answering have become more accurate and widely applicable.
BERT is one of the most well-known models in NLP and has been applied to a variety of tasks.
"""

# List of questions
questions = [
    "What happened in 2024?",
    "What tasks are transformer-based models used for?",
    "Which model is well-known in NLP?"
]

# Iterate through the list of questions and extract answers
for question in questions:
    # Tokenize question and context
    inputs = tokenizer(question, context, return_tensors='pt')

    # Perform inference
    with torch.no_grad():
        outputs = model(**inputs)

    # Get the most likely start and end of the answer
    answer_start_index = torch.argmax(outputs.start_logits)
    answer_end_index = torch.argmax(outputs.end_logits) + 1

    # Convert tokens to string (the answer)
    answer = tokenizer.convert_tokens_to_string(tokenizer.convert_ids_to_tokens(inputs['input_ids'][0][answer_start_index:answer_end_index]))

    # Print the question and answer
    print(f"Question: {question}")
    print(f"Answer: {answer}\n")
```
